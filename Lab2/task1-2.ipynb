```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
img_path = "img2.jpg"
im = Image.open(img_path)
img_arr = np.array(im)
print(img_arr.shape)
```

# Task 1

```python
import matplotlib.image as mpimg

def gray1(rgb):
    return np.dot(rgb[..., :3], [0.2989, 0.5870, 0.1140]).astype(int)

def gray2(rgb):
    return np.dot(rgb[..., :3], [0.2126, 0.7152, 0.0722]).astype(int)

def getdiff(g1, g2):
    return (abs(g1 - g2)).clip(0,255) 

gray11 = gray1(img_arr)
gray22 = gray2(img_arr)
diff = np.stack([getdiff(gray11, gray22)] * 3, axis=-1)
plt.imshow(gray11, cmap=plt.get_cmap('gray'))
plt.show()
plt.imshow(gray22, cmap=plt.get_cmap('gray'))
plt.show()
plt.imshow(diff, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
fig = plt.figure(figsize = (15, 10))

ax = fig.add_subplot(2, 1, 1)
plt.title('HIST 1')
plt.hist(gray11.ravel(), bins=256, color='k')
ax = fig.add_subplot(2, 1, 2)
plt.title('HIST 2')
plt.hist(gray22.ravel(), bins=256, color='k')
plt.show()
```

```python
fig = plt.figure(figsize = (15, 10))

ax = fig.add_subplot(3, 2, 1)
plt.title('RED')
plt.imshow(img_arr[:,:,0], plt.cm.Reds_r)
plt.axis('off')
ax = fig.add_subplot(3, 2, 3)
plt.title('GREEN')
plt.imshow(img_arr[:,:,1], plt.cm.Greens_r)
plt.axis('off')
ax = fig.add_subplot(3, 2, 4)
plt.title('BLUE')
plt.imshow(img_arr[:,:,2], plt.cm.Blues_r)
plt.axis('off')
ax = fig.add_subplot(3, 2, 2)
plt.title('ORIGINAL')
plt.imshow(img_arr)
plt.axis('off')
ax = fig.add_subplot(3, 1, 3)
plt.title('HIST')
plt.hist(img_arr[:,:,0].ravel(), bins=256, color = 'r', alpha=0.3)
plt.hist(img_arr[:,:,1].ravel(), bins=256, color = 'g', alpha=0.3)
plt.hist(img_arr[:,:,2].ravel(), bins=256, color = 'b', alpha=0.3)
plt.tight_layout()
plt.show()
```
